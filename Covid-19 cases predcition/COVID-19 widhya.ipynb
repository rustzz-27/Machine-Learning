```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
df=pd.read_csv('covid19.csv')
df.head()
```

```python
df.drop('Sno',axis=1,inplace=True)
```

```python
cd=df.groupby(['Date'],sort=False).sum()
```

```python
cd.head(100)
```

```python
fd=cd.reset_index()
```

```python
cd['Total']=0
```

```python
fd
```

```python
for i in range(len(fd)):
    x=np.sum(cd.iloc[i])
    fd.loc[i,'Total']= x
```

```python
fd
```

```python
fd=fd.astype({'Total':int})
```

```python
fd.plot(kind='bar',x='Date',y='Total',figsize=(20,5),grid=True,title="Dates VS Total Cases")
```

```python
from math import e
def expo(P_o,t):
    r=0
    count=0
    for i in range(34,len(fd)-1,1):
        count+=1
        r=r+((fd.loc[i+1,'Total']-fd.loc[i,'Total'])//fd.loc[i,'Total'])
    r=r//count
    P_t=P_o*(e**(r*t))
    print("Predicted Forecast for 04/03/20:",int(P_t),"cases")
```

```python
if __name__ == '__main__':
    expo(31,26)
```
